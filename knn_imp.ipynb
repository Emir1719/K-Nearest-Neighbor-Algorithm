```python
import math
import pandas as pd
```

```python
# Uzaklık hesabı için öklid fonksiyonu:

def oklid_distance(data1 :list, data2 :list):
  lenght = len(data1)
  if lenght != len(data2):
    raise ValueError("Veri noktalarının uzunlukları eşit olmalı.")

  sum = 0
  for i in range(lenght):
    sum += (data1[i] - data2[i])**2

  return math.sqrt(sum)
```

```python
# Veri setinden alınan veriler listeye dönüştürülür:

data = pd.read_csv("diabetes.csv")
data2 = data.drop(["Outcome"], axis=1) # Bağımlı değişken kaldırıldı
list = data2.values.tolist()
list[0]
```

```python
# Öklid fonksiyonun testi

print(oklid_distance(list[0], list[0]))
print(oklid_distance(list[0], list[1]))
print(oklid_distance(list[0], list[2]))
print(oklid_distance(list[0], list[3]))
```

```python
# Yeni istenen [example] verisinin listedeki en yakın olanları 
# bulunur ve küçükten büyüğe listelenir: 
data_new = None

def listToDataframe(array):
    global data_new
    list_sum = []

    for i in range(0, len(list)):
        list_sum.append(oklid_distance(array, list[i]))

    data_new = pd.DataFrame(list_sum, columns=["Distance"])
    data_new = pd.concat([data_new, data], axis=1)
    data_new = data_new.sort_values(["Distance"])
    return data_new
```

```python
def knn(array):
    global data_new
    k = 5
    data_new = listToDataframe(array)
    # İlk k satırdaki "Outcome" alanını listeye atma
    outcome_list = data_new.head(k).loc[:, "Outcome"].tolist()

    # 1: hasta, 0: sağlıklı
    hasta = 0
    saglikli = 0

    for i in range(0, k):
        number = outcome_list[i]
        if number == 0:
            saglikli = saglikli + 1
        else:
            hasta = hasta + 1

    if saglikli > hasta:
        print("Kişi sağlıklı")
    else:
        print("Kişi hasta")

    # Listeyi yazdırma
    print(F"İlk {k} verideki sonuçlar", outcome_list,"\n")
    data_new
```
